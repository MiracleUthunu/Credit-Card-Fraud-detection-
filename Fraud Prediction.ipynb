```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
fraud.columns
```

```python
fraud.info()
```

```python
fraud['merchants'] = encoder.fit_transform(fraud['merchant'])
fraud['categorys'] = encoder.fit_transform(fraud['category'])
fraud['DOB'] = encoder.fit_transform(fraud['dob'])
fraud['JOB'] = encoder.fit_transform(fraud['job'])
fraud['trans_date'] = encoder.fit_transform(fraud['trans_date_trans_time'])
fraud['trans_details'] = encoder.fit_transform(fraud['trans_num'])
```

```python
fraud.drop(columns='merchant', axis=1, inplace=True)
fraud.drop(columns='category', axis=1, inplace=True)
fraud.drop(columns='dob', axis=1, inplace=True)
fraud.drop(columns='job', axis=1, inplace=True)
fraud.drop(columns='trans_num', axis=1, inplace=True)
fraud.drop(columns='trans_date_trans_time', axis=1, inplace=True)

fraud.head()
```

```python
fraud.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
x = fraud.drop(columns=['is_fraud', 'state', 'street', 'city', 'gender', 'last', 'first', 'unix_time', 'merch_lat','merch_long'])
y = fraud['is_fraud']
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score, mean_absolute_error
```

```python
print("Acurracy:", accuracy_score(y_test, y_pred))
```

```python
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
```

```python
print("Precision:", precision_score(y_test, y_pred))
```

```python

min_length_train = min(len(x_train), len(y_train))
x_train = x_train[:min_length_train]
y_train = y_train[:min_length_train]

min_length_test = min(len(x_test), len(y_test))
x_test = x_test[:min_length_test]
y_test = y_test[:min_length_test]
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("MAE:" , mae)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
print(y_pred)
```

```python
print("Acurracy:", accuracy_score(y_test, y_pred))
```

```python
print("Precision:", precision_score(y_test, y_pred))
```
